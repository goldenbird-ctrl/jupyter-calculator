```python
def add(x, y): 
    print("Result:", x + y)

def subtract(x, y): 
    print("Result:", x - y)

def multiply(x, y): 
    print("Result:", x * y)

def divide(x, y): 
    if y != 0:
        print("Result:", x / y)
    else:
        print("Error! division by zero.")

def calculator():
    print("---Jupyter Notebook Calculator---")
    print("Operation: + , - , * , /")
    print("Type 'exit' at any time to quit.")
    
    while True:
        user_input = input("\nEnter first number (or 'exit'): ").strip()
        if user_input.lower() == 'exit':
            print("Calculator closed. Goodbye!")
            break
            
        operation = input("Enter operation (+, -, *, /): ").strip()
        if operation.lower() == 'exit':
            print("Calculator closed. Goodbye!")
            break
            
        next_input = input("Enter second number: ").strip()
        if next_input.lower() == 'exit':
            print("Calculator closed. Goodbye!")
            break 
            
        try:
            num1 = float(user_input)
            num2 = float(next_input)
            
            if operation == '+':
                add(num1, num2)
            elif operation == '-':
                subtract(num1, num2)
            elif operation == '*':
                multiply(num1, num2)
            elif operation == '/':
                divide(num1, num2)
            else:
                print("Invalid operation! Please use +, -, *, or /")
                
        except ValueError:
            print("Please enter valid numbers.")
calculator()
```
